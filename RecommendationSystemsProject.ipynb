                    3. Build a popularity based model and recommend top 5 mobile phones. 

Popularity based recommendation system works with the trend. It basically recommends the items which are in trend right now. There is no point of personalization to user here because we sort out depends on popularity and this helps us a lot for new users.
The most popular one will recommend to a user bcz it is the most viewed or solded item as per time being....

```python
ratings=pd.DataFrame(newdf,columns=['score','product'])
```

```python
ratings.sort_values('score',ascending=False).head(5)
```

```python
#we have a lot of highest rated products are there, lets work on how many times it has rated and average of rating also
```

```python
#By average of score rating...
meanrating=newdf.groupby('product')['score'].mean().sort_values(ascending = False).head(5)
```

```python
print('The top 5 recommened products based on average rating are.............\n ')
for i in meanrating.index:
    print(i)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(5,5))
plt.barh(meanrating.index,meanrating,align='center',color='skyblue')
plt.xlabel('Rating')
plt.title('Popularity based recommended the top 5 phones on average rating of the product\n')
```

```python
#By count of no of times the product rated
Countrating=newdf.groupby('product')['score'].count().sort_values(ascending = False).head(5)
```

```python
print('The top 5 recommened products based on no of times the product rated are.............\n ')
for i in Countrating.index:
    print(i)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(5,5))
plt.barh(Countrating.index,Countrating,align='center',color='skyblue')
plt.xlabel('Rating count')
plt.title('Popularity based recommended the top 5 phones on Counting the times that product rated.\n')
```

4. Build a collaborative filtering model using SVD. You can use SVD from surprise or build it from scratch

Build a collaborative filtering model using kNNWithMeans from surprise. Youcan try both user-based and item-based model.

```python
from surprise import KNNWithMeans as knn
from surprise.model_selection import train_test_split
from surprise import accuracy
from surprise import SVD
from surprise import Reader
from surprise import Dataset as ds
```

```python
newdf.head(2)
```

```python
#converting the ratings to a scale of 1 to 10 using Reader here...
reader = Reader(rating_scale=(0, 10))

#here there is no need of country and source columns for making score prediction
dataset=ds.load_from_df(newdf[['author','product','score']],reader)
dataset
```

```python
trainset,testset=train_test_split(dataset,test_size=0.3)
```

USER BASED collaborative filtering 

```python
ub=knn(k=50,verbose=True,sim_options={'name':'pearson_baseline','user_based':True})
```

```python
ub.fit(trainset)
```

```python
testpred=ub.test(testset)
```

```python
testpred
```

```python
userrmse=accuracy.rmse(testpred)
```

```python
from surprise.model_selection import KFold
kf=KFold()
```

```python
for training_data, testing_data in kf.split(dataset):
    model.fit(training_data)
    predictions = model.test(testing_data)
    accuracy.rmse(predictions)
```
